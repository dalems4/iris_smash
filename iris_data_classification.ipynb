```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import ensemble
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)
```

```python
gbc = ensemble.GradientBoostingClassifier(learning_rate=.01, max_depth=3,n_estimators=1000)
```

```python
gbc.fit(X_train, y_train)
```

```python
predictions = gbc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, predictions))
```

```python
def byte_to_json(byte_content):
    json = byte_content.content.decode('utf8').replace("'", '"')
    
    # Load the JSON to a Python list & dump it back out as formatted JSON
    data = json.loads(json)
    formatted_json = json.dumps(data, indent=4, sort_keys=True)
    return formatted_json
```
